```python
import os 
import pandas as pd

os.getcwd()
```

```python
# Load both datasets
hq = pd.read_csv("hq/high_quality_manga_sales.csv")
pop = pd.read_csv("popular/popular_manga_sales.csv")

# Union them
combined = pd.concat([hq, pop], ignore_index=True)

# Check duplicates
print(combined.duplicated(subset=["manga_id"]).sum())

combined = combined.sort_values(by="sales_circulation", na_position="last")

# Drop duplicates
combined = combined.drop_duplicates(subset=["manga_id"],keep="first")
```
